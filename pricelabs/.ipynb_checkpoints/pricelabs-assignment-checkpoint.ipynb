```python
import warnings
warnings.filterwarnings('ignore')
import math

import matplotlib.pyplot as plt 
import numpy as np 
import os 
import pandas as pd 
from datetime import datetime
%matplotlib inline
from matplotlib.dates import DateFormatter
date_form = DateFormatter("%b,%y")
from matplotlib import rcParams
import matplotlib.ticker as mtick
rcParams['figure.figsize'] = (8, 3)
rcParams['figure.dpi'] = 150
rcParams['lines.linewidth'] = 2

import seaborn as sns
sns.set_style("darkgrid")

from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
def plot_chart(title='', title_size=40,
               ylabel='', ylabel_size=10, yticks_size=10, yticks_rotation=0,
               xlabel='', xlabel_size=10, xticks_size=10, xticks_rotation=0, xticks_labels=None,
               legend=False, legend_size=15, legend_loc='best', legend_ncol=1):
    
    plt.title(title, fontsize=title_size)
    
    plt.xlabel(xlabel, fontsize=xlabel_size)
    if (xticks_labels):
        plt.xticks(xticks_labels, fontsize=xticks_size, rotation=xticks_rotation)   
    else:
        plt.xticks(fontsize=xticks_size, rotation=xticks_rotation)
    
    plt.ylabel(ylabel, fontsize=ylabel_size)
    plt.yticks(fontsize=yticks_size, rotation=yticks_rotation)
    
    if (legend):
        plt.legend(prop={'size': legend_size}, loc=legend_loc, ncol=legend_ncol)
    plt.show()
    
```

```python
def testing_stationarity(dataframe, window = 365):
    dataframe["Rolling_Mean"] = dataframe["Price"].rolling(window= window).mean()
    dataframe["Rolling_Std"] = dataframe["Price"].rolling(window= window).std()
    dataframe = dataframe[["Date", "Price", "Rolling_Mean", "Rolling_Std"]]
    ax = dataframe.set_index("Date").plot()
    ax.xaxis.set_major_formatter(date_form)
    plot_chart(title='Rolling Mean & Standard Deviation', title_size=16,legend=True, legend_size=10)
    
    #Perform Augmented Dickeyâ€“Fuller test:
    print('Results of Dickey Fuller Test:')
    dftest = adfuller(dataframe["Price"], autolag='AIC')

    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
    
```

```python
def best_fit(X, Y):

    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum(xi*yi for xi,yi in zip(X, Y)) - n * xbar * ybar
    denum = sum(xi**2 for xi in X) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.2f} + {:.2f}x'.format(a, b))

    return a, b
```

```python
def model_performance(y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    print('MSE: '+str(mse))
    mae = mean_absolute_error(y_true, y_pred)
    print('MAE: '+str(mae))
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    print('RMSE: '+str(rmse))
    mape = np.mean(np.abs(y_pred - y_true)/np.abs(y_true))
    print('MAPE: '+str(mape))
```

# Baseline Approach
*Step1* -> Made time series stationary by subtracting each data point by mean price value of that particular month.<br>
*Step2* -> From above stationary time series data we took mean of feb month data for year (2012,2013, 2014, 2015).<br>
*Step3* -> From orginal data trend we predicted Mean price of Feb 2020.<br>
*Step4* -> Took step2 prediction for feb month and added predicted mean price of Feb 2020 to get final prediction price.

```python
## mean df 
mean_df = df.groupby(["year", "month"])["Price"].mean().reset_index()
mean_df = mean_df[~mean_df["Price"].isnull()]
mean_df.rename(columns={"Price":"meanprice"},inplace=True)

## joining
stationary_df = df.merge(mean_df, on=["year", "month"],how="left")
stationary_df["price_meanprice"] = stationary_df["Price"] - stationary_df["meanprice"]

stationary_df.set_index("Date")["price_meanprice"].plot()
plot_chart(ylabel="Price", legend=False)
```

```python
## filter: feb
stationary_feb_df = stationary_df[stationary_df["month"] == "Feb"]

stationary_feb_df = stationary_feb_df.pivot_table(index="day", columns="year", values="price_meanprice")
stationary_feb_df["Mean"] = stationary_feb_df.mean(axis=1)

ax = stationary_feb_df[["Mean"]].plot(style='o-', linewidth=3.5)
stationary_feb_df[[2012, 2013, 2014, 2015]].plot(ax=ax, style='.-', linewidth=1, alpha=0.4)

plt.axhline(0, color="#ffffff")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
loc = ax.xaxis.get_major_locator();loc.set_params(integer=True)
plot_chart(ylabel="Price", xlabel="Day", legend=False)
```

```python
## agg
year_df = df[df["year"] <= 2015].groupby("year").agg({"Price":"mean"})
jan_df = df[df["month"] == "Jan"].groupby("year").agg({"Price":"mean"})
feb_df = df[df["month"] == "Feb"].groupby("year").agg({"Price":"mean"})

## best fit
X1, Y1 = year_df.index, year_df["Price"]
a1, b1 = best_fit(X1, Y1)
yfit1 = [a1+b1 * xi for xi in X1]

X2, Y2= feb_df.index, feb_df["Price"]
a2, b2 = best_fit(X2, Y2)
yfit2 = [a2+b2 * xi for xi in X2]

X3, Y3= jan_df.index, jan_df["Price"]
a3, b3 = best_fit(X3, Y3)
yfit3 = [a3+b3 * xi for xi in X3]

## predictions
xpred = [2015, 2016, 2017, 2018, 2019, 2020]
ypred1 = [a1 + b1 * xi for xi in xpred]
ypred2 = [a2 + b2 * xi for xi in xpred]
ypred3 = [a3 + b3 * xi for xi in xpred]

##
ax = sns.scatterplot(x=X1, y=Y1, s=50, color="tab:blue")
sns.scatterplot(x=X3, y=Y3, ax=ax, s=50, color="tab:orange")
sns.scatterplot(x=X2, y=Y2, ax=ax, s=50, color="tab:green")

plt.plot(X1, yfit1,label="Overall ({} + {}x)".format(round(a1, 1), round(b1, 1)))
plt.plot(xpred, ypred1, "--", color="tab:blue", marker=".")

plt.plot(X3, yfit3, label="January ({} + {}x)".format(round(a3, 1), round(b3, 1)))
plt.plot(xpred, ypred3, "--", color="tab:orange", marker=".")

plt.plot(X2, yfit2, label="February ({} + {}x)".format(round(a2, 1), round(b2, 1)))
plt.plot(xpred, ypred2, "--", color="tab:green", marker=".")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=12)
loc = ax.xaxis.get_major_locator();loc.set_params(integer=True)
plot_chart(ylabel="Price", legend=False)
```

# Double Exponential Smoothing
*Step1* -> Determined optimal alpha, gamma using grid search<br>
*Step2* -> Forecasted next 45 days<br>
We see a saturation in out of sample forecast curve using Double Exponential Smoothing. Hence we didn't used this approach for final predictions as it is only good for predicting 5-6 days in future.


```python
#Double Exponential Smoothing
optimal_alpha = None
optimal_gamma = None
best_mse = None
csv_dataset = df[['Price']].copy()
db = csv_dataset.iloc[:, :].values.astype('float32')
db_train = csv_dataset.iloc[:1095, :].values.astype('float32')
db_valid = csv_dataset.iloc[1095:, :].values.astype('float32')
mean_results_for_all_possible_alpha_gamma_values = np.zeros((9, 9))
mean_results_for_all_possible_alpha_gamma_values_valid = np.zeros((9, 9))
for gamma in range(0, 9):
    for alpha in range(0, 9):
        pt = db_train[0][0]
        bt = db_train[1][0] - db_train[0][0]
        mean_for_alpha_gamma = np.zeros(len(db_train))
        mean_for_alpha_gamma[0] = np.power(db_train[0][0] - pt, 2)
        for i in range(1, len(db_train)):
            temp_pt = ((alpha + 1) * 0.1) * db_train[i][0] + (1 - ((alpha + 1) * 0.1)) * (pt + bt)
            bt = ((gamma + 1) * 0.1) * (temp_pt - pt) + (1 - ((gamma + 1) * 0.1)) * bt
            pt = temp_pt
            mean_for_alpha_gamma[i] = np.power(db_train[i][0] - pt, 2)
        mean_results_for_all_possible_alpha_gamma_values[gamma][alpha] = np.mean(mean_for_alpha_gamma)
        
        pt = pt
        bt = bt
        mean_for_alpha_gamma_valid = np.zeros(len(db_valid))
        mean_for_alpha_gamma_valid[0] = np.power(db_valid[0][0] - pt, 2)
        for i in range(1, len(db_valid)):
            temp_pt = ((alpha + 1) * 0.1) * db_valid[i][0] + (1 - ((alpha + 1) * 0.1)) * (pt + bt)
            bt = ((gamma + 1) * 0.1) * (temp_pt - pt) + (1 - ((gamma + 1) * 0.1)) * bt
            pt = temp_pt
            mean_for_alpha_gamma[i] = np.power(db_valid[i][0] - pt, 2)
        mean_results_for_all_possible_alpha_gamma_values_valid[gamma][alpha] = np.mean(mean_for_alpha_gamma_valid)
        
        optimal_gamma, optimal_alpha = np.unravel_index(
            np.argmin(mean_results_for_all_possible_alpha_gamma_values_valid),
            np.shape(mean_results_for_all_possible_alpha_gamma_values_valid))
optimal_alpha = (optimal_alpha + 1) * 0.1
optimal_gamma = (optimal_gamma + 1) * 0.1
best_mse = np.min(mean_results_for_all_possible_alpha_gamma_values)
print("Best MSE = %s" % best_mse)
print("Optimal alpha = %s" % optimal_alpha)
print("Optimal gamma = %s" % optimal_gamma)
```

```python
forecast = np.zeros(len(db) + 45)
pt = db[0][0]
bt = db[1][0] - db[0][0]
forecast[0] = pt
for i in range(1, len(forecast)):
    temp_pt = optimal_alpha * db[i][0] + (1 - optimal_alpha) * (pt + bt)
    bt = optimal_gamma * (temp_pt - pt) + (1 - optimal_gamma) * bt
    pt = temp_pt
    forecast[i] = pt
    if i >= 1475:
        db = np.append(db, [[pt]], axis = 0)
forecast[-1] = pt + (1 * bt)
plt.plot(db[1095:, 0],label = 'real data')
plt.plot(forecast[1095:], label = 'forecast')
plt.legend()
plt.show()
```

# SARIMAX (ARIMA with Exogenous Features)
Step1 -> Created rolling features of mean and std with window = (30,45,60,90,120,180,365) <br>
Step2 -> Added day, week, month, dayofweek features <br>
Step3 -> removed outliers <br>
Step4 -> Split data into train-valid in 75-25 split <br>
Step5 -> Used Auto ARIMA to find and train on optimal hyperparameters.<br>
Step6 -> Forecasted data for Feb 2016 by creating rolling features from last prediction.<br>
Step7 -> Adjusted Feb 2016 predicted price to Feb 2020 by adding mean_diff to Feb 2016 predicted price. Mean_diff calculated by difference linear regression prediction of (mean of Feb 2020) - (mean of Feb 2016)

```python
sarimax_df = df.copy()
windows_list = [30,45,60,90,120,180,365]
sarimax_df = sarimax_df.drop(index = sarimax_df[['Price']].idxmax())

for window in windows_list:
    temp_rolled_d = sarimax_df["Price"].rolling(window=window, min_periods=0)
    temp_mean_d = temp_rolled_d.mean().shift(1).reset_index().astype(np.float32)
    temp_std_d = temp_rolled_d.std().shift(1).reset_index().astype(np.float32)
    sarimax_df[f"Price_mean_lag{window}"] = temp_mean_d["Price"]
    sarimax_df[f"Price_std_lag{window}"] = temp_std_d["Price"]
    
sarimax_df.fillna(sarimax_df.mean(), inplace=True)
sarimax_df.set_index("Date", drop=False, inplace=True)
sarimax_df.Date = pd.to_datetime(sarimax_df.Date, format="%Y-%m-%d")
sarimax_df["month"] = sarimax_df.Date.dt.month
sarimax_df["week"] = sarimax_df.Date.dt.week
sarimax_df["day"] = sarimax_df.Date.dt.day
sarimax_df["day_of_week"] = sarimax_df.Date.dt.dayofweek
sarimax_df.head()
```

```python
df_train = sarimax_df[sarimax_df.Date < "2015"]
df_valid = sarimax_df[sarimax_df.Date >= "2015"]

exogenous_features = ['month', 'year', 'Price_mean_lag30','Price_mean_lag45',
                      'Price_mean_lag60', 'Price_mean_lag90',
                       'Price_mean_lag120', 'Price_mean_lag180', 'Price_mean_lag365',
                      'Price_std_lag30', 'Price_std_lag45', 'Price_std_lag60',
                   'Price_std_lag90', 'Price_std_lag120', 'Price_std_lag180',
                   'Price_std_lag365', 'week', 'day', 'day_of_week']
```

```python
model = auto_arima(df_train.Price, exogenous=df_train[exogenous_features], trace=True, error_action="ignore", suppress_warnings=True)
model.fit(df_train.Price, exogenous=df_train[exogenous_features])
```

```python
forecast = model.predict(n_periods=len(df_valid), exogenous=df_valid[exogenous_features])
df_valid["Forecast_ARIMAX"] = forecast
df_valid[["Price", "Forecast_ARIMAX"]].plot(figsize=(14, 7))
```

```python
model_performance(df_valid["Price"], df_valid["Forecast_ARIMAX"])
```

```python
forecast_steps = 90
for i in range(forecast_steps):
    last_date_index = df_valid.iloc[[-1]].index + pd.Timedelta(days=1)
    last_date = pd.to_datetime(last_date_index).date[0]
    last_time = pd.to_datetime(pd.to_datetime(last_date_index).date[0], format="%Y-%m-%d")
    data = { 'Date' : last_date,
            'month' : last_date.month,
            'year': last_date.year, 
            'week' : last_time.week,
            'day': last_date.day,
            'day_of_week' : last_time.dayofweek,
    }
    df_valid = df_valid.append(pd.DataFrame(data, index=last_date_index))
    for window in windows_list:
        temp_rolled_d = df_valid["Price"].rolling(window=window, min_periods=0)
        temp_mean_d = temp_rolled_d.mean().shift(1).to_frame()
        temp_std_d = temp_rolled_d.std().shift(1).to_frame()
        df_valid.at[last_date_index, f'Price_mean_lag{window}']  = temp_mean_d.loc[last_date_index]["Price"]
        df_valid.at[last_date_index, f'Price_std_lag{window}']  = temp_std_d.loc[last_date_index]["Price"]

    forecast = model.predict(n_periods=1, exogenous=df_valid.loc[last_date_index][exogenous_features])
    df_valid.at[last_date_index, 'Forecast_ARIMAX']  = forecast
    df_valid.at[last_date_index, 'Price']  = forecast
    
```
